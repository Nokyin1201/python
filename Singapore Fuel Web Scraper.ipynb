```python
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import re
import pandas as pd

driver = webdriver.Chrome("C:/Users/123/Desktop/Python Project/chromedriver.exe")
driver.get("https://fuelkaki.sg/home")

time.sleep(5)  
soup = BeautifulSoup(driver.page_source, 'html.parser')
#print(soup)
```

```python
fuel_name = soup.find_all("div", {'class': 'fuel-name'}) 
#print(fuel_name)
Fuel_namelist = []                  
for Fuel_name1 in fuel_name:
    FuelName = Fuel_name1.text
    Fuel_namelist.append(FuelName)
new_Fuel_List = Fuel_namelist[0:5]
print(new_Fuel_List)   
```

```python
time = soup.find_all("div", {'class': 'time'})       
#print(time)

#Caltex_date = time[0].text # string
#print(Caltex_date)

#Esso_date = time[1].text
#print(Esso_date)

#Shell_date = time[2].text
#print(Shell_date)

#Sinopec_date = time[3].text
#print(Sinopec_date)

#SPC_date = time[4].text
#print(SPC_date)

#date_lists = [Caltex_date]
#for date_list in date_lists:
#    print(date_list)
#new_date_list = [Caltex_date.replace('\u202c','') for date_list in date_lists]

#print(new_date_list)

Time_List = []
for Time1 in time:
    #print(Time1.text)
    Time1 = Time1.text
    Time_List.append(Time1)
Time_List = Time_List[0:5]
#print(Time_List)

new_Time_List = []
for Time_Lists in Time_List:
    new_Time_List += [Time_Lists.replace('\u202c', '')]

print(new_Time_List)

price = soup.find_all("td")
#print(price)
```

```python
Caltex_price = price[1:6]   
#print(Caltex_price)

Caltex = []
for Caltex_Price1 in Caltex_price:
    CaltexPrice = Caltex_Price1.text
    Caltex.append(CaltexPrice)            
#print(Caltex)

new_Caltex_list = []
for Caltexs in Caltex:
    new_Caltex_list += [Caltexs.replace('\u202c', '').replace('S$ ','').replace('(Platinum 98)', '')]  # 移除多餘的文字
print("Caltex: ")
print(new_Caltex_list)

Esso_price = price[7:12]
#print(Esso_price)

Esso = []
for Esso_price1 in Esso_price:
    EssoPrice = Esso_price1.text
    Esso.append(EssoPrice)
#print(Esso)

new_Esso_list = []
for Essos in Esso:
    new_Esso_list += [Essos.replace('\u202c', '').replace('S$ ','').replace('Diesel', '').replace('Regular', '').replace('Extra', '').replace('(Synergy Supreme+)', '')]
print("Esso: ")
print(new_Esso_list)

Shell_price = price[13:18]
#print(Shell_price)

Shell = []
for Shell_Price1 in Shell_price:
    ShellPrice = Shell_Price1.text
    Shell.append(ShellPrice)
#print(Shell)

new_Shell_list = []
for Shells in Shell:
    new_Shell_list += [Shells.replace('\u202c', '').replace('S$ ','').replace('(Shell V-Power)', '')]
print("Shell: ")
print(new_Shell_list)

Sinopec_price = price[19:24]

Sinopec = []
for Sinopec_Price1 in Sinopec_price:
    SinopecPrice = Sinopec_Price1.text
    Sinopec.append(SinopecPrice)
    
new_Sinopec_list = []
for Sinopecs in Sinopec:
    new_Sinopec_list += [Sinopecs.replace('\u202c', '').replace('S$ ','').replace('(SINO X Power)', '')]
print("Sinopec: ")
print(new_Sinopec_list)

SPC_price = price[25:30]
SPC = []
for SPC_Price1 in SPC_price:
    SPCPrice = SPC_Price1.text
    SPC.append(SPCPrice)

new_SPC_list = []
for SPCs in SPC:
    new_SPC_list += [SPCs.replace('\u202c', '').replace('S$ ','')]
print("SPC")
print(new_SPC_list)

Diesel = [new_Caltex_list[0], new_Esso_list[0], new_Shell_list[0], new_Sinopec_list[0], new_SPC_list[0]]
print("Diesel: ")
print(Diesel)

P92 = [new_Caltex_list[1], new_Esso_list[1], new_Shell_list[1], new_Sinopec_list[1], new_SPC_list[1]]
print("P92: ")
print(P92)

P95 = [new_Caltex_list[2], new_Esso_list[2], new_Shell_list[2], new_Sinopec_list[2], new_SPC_list[2]]
print("P95: ")
print(P95)

P98 = [new_Caltex_list[3], new_Esso_list[3], new_Shell_list[3], new_Sinopec_list[3], new_SPC_list[3]]
print("P98: ")
print(P98)

Premium = [new_Caltex_list[4], new_Esso_list[4], new_Shell_list[4], new_Sinopec_list[4], new_SPC_list[4]]
print("Premium: ")
print(Premium)

driver.close()
```
