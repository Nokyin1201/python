```python
from urllib.request import urlopen
import pandas as pd
import requests
from bs4 import BeautifulSoup
import datetime
```

```python
consumer_council_D = requests.get("https://oil-price.consumer.org.hk/tc/diesel")
soup_C_D = BeautifulSoup(consumer_council_D.text, "html.parser")
print(soup_C_D.prettify())
```

```python
name2 = soup_C_D.find_all("span", {'class': 'clogo__name'})
#print(name2)
N_list = []
for N2 in name2:
    #print(N2.text)
    N4 = N2.text
    N_list.append(N4)
print(N_list)    
```

```python
D_list = []
D = soup_C_D.find_all("div", {'class': 'board__cell board__cell--2'})
#print(D)
for C_D in D:
    #print(C_D.text)
    D4 = C_D.text.strip()
    D_list.append(D4)
#print(D_list)
del D_list[0]

# create a dataframe to store above two list
df_Diesel = pd.DataFrame(list(zip(N_list, D_list)), columns = ['Name', 'Diesel Price'])
df_Diesel
```

```python
consumer_council = requests.get("https://oil-price.consumer.org.hk/tc")
soup_consumer = BeautifulSoup(consumer_council.text, "html.parser")
#print(soup_consumer.prettify())
```

```python
result_consumer_date = soup_consumer.find_all('p')
print("消委會網站: " + str(result_consumer_date[4].get_text()))
```

```python
consumer_P = soup_consumer.find_all("div", {'class': 'board__cell board__cell--2'})
name1 = soup_consumer.find_all("span", {'class': 'clogo__name'})
#print(name1)
n_lst = []
for N in name1:
    n_1 = N.text
    n_lst.append(n_1)
#print(n_lst)
n_lst_copy = n_lst[:]
#print(n_lst_copy)

del n_lst[5:]
#print(n_lst)
del n_lst_copy[0:5]
#print(n_lst_copy)

C_P_lst = []
#print(consumer_P)
for C_P in consumer_P:
    #print(C_P.text)
    C_P_1 = C_P.text.strip()
    C_P_lst.append(C_P_1)
#print(C_P_lst)
C_P_lst_super = C_P_lst[:]
#print(C_P_lst_super)

del C_P_lst[0]
del C_P_lst[5:]
#print(C_P_lst)

del C_P_lst_super[0:7]
#print(C_P_lst_super)
```

```python
df_Petro = pd.DataFrame(list(zip(n_lst, C_P_lst)), columns = ['Name', 'Petro Price'])
df_Petro
```

```python
df_SuperPetro = pd.DataFrame(list(zip(n_lst_copy, C_P_lst_super)), columns = ['Name', 'SuperPetro Price'])
df_SuperPetro
```

```python
Webdate = "消委會網站: " + str(result_consumer_date[4].get_text())
print(Webdate)
frames = [df_Diesel, df_Petro, df_SuperPetro]
combine = pd.concat(frames, axis=1)
combine
combine=combine.set_axis(['柴油','Diesel Price','汽油(平)','Petro Price','汽油(貴)','SuperPetro Price'],axis=1)
#print(combine)
```

```python
from tkinter import *  
from tkinter import ttk
root = Tk()  

cols = list(combine.columns)

tree = ttk.Treeview(root)
tree.pack()
tree["columns"] = cols
for i in cols:
    tree.column(i, anchor="w")
    tree.heading(i, text=i, anchor='w')

for index, row in combine.iterrows():
    tree.insert("",0,text=index,values=list(row))

root.title(Webdate)
exit_button = Button(root, text="OK!", command=root.destroy)
exit_button.pack(pady=20)
root.mainloop()
```
